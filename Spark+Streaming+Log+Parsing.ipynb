```python
from pyspark.sql import SparkSession, Row
from pyspark.streaming import StreamingContext
from pyspark.sql.functions import regexp_extract
```

```python
spark = SparkSession.builder.appName('logparserstream').getOrCreate()
sc = spark.sparkContext
ssc = StreamingContext(sc,2)
```

```python
filepath = '/user/edureka_1033463/streaming'
filepath
```

```python
line = ssc.textFileStream(filepath)
line
```

```python
def process(t, rdd):
        if rdd.isEmpty():
            print("==== EMPTY ====")
            return

        print("=== RDD Found ===")
        rowRdd = rdd.map(lambda x: Row(line=x))
        spark = SparkSession.builder.appName("logparserstream").getOrCreate()
        df = spark.createDataFrame(rowRdd)
        print(df.show())
        
        host_pattern = r'(^\S+\.[\S+\.]+\S+)\s'
        ts_pattern = r'\[(\d{2}/\w{3}/\d{4}:\d{2}:\d{2}:\d{2} -\d{4})]'
        method_uri_protocol_pattern = r'\"(\S+)\s(\S+)\s*(\S*)\"'
        status_pattern = r'\s(\d{3})\s'
        content_size_pattern = r'\s(\d+)$'
        
        df2 = df.select(regexp_extract('line', host_pattern, 1).alias("host"),
                        regexp_extract('line', ts_pattern, 1).alias("timestamp"),
                        regexp_extract('line', method_uri_protocol_pattern, 1).alias("method"),
                        regexp_extract('line', method_uri_protocol_pattern, 2).alias("endpoint"),
                        regexp_extract('line', method_uri_protocol_pattern, 3).alias("protocol"),
                        regexp_extract('line', status_pattern, 1).alias("status"),
                        regexp_extract('line', content_size_pattern, 1).alias("content_size")
                       )

        df2 = df2.filter("status != ''")
        print(df2.show()) 
        df2.coalesce(1).write.format("text").option("header", "true").csv("hdfs:///user/edureka_1033463/created/dataframe")
```

```python
line.pprint()
line.foreachRDD(process)
```

```python
ssc.start()
ssc.awaitTermination()
```

```python
ssc.stop()
```
